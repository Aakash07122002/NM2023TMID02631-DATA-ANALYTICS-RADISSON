#**House Price India Visualization**

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 190, where=(ys > 190), facecolor='r', alpha=0.6)

plt.title("House Price India")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('House Price India.csv')
df.head()
```

```python
df.info()
```

# **Univariate Analysis**

##Histogram

```python
plt.hist(df['number of bedrooms'],bins=50)
plt.xlabel("No.of.bedrooms")
plt.ylabel("Count")
```

##Displot

```python
sns.distplot(df['number of floors'],bins=30)
```

##Boxplot

```python
sns.boxplot(df['Price'])
```

##Violinplot

```python
sns.violinplot(x=df['lot_area_renov'])
```

# **Bivariate Analysis**

##Scatterplot

```python
sns.scatterplot(x=df['number of bedrooms'],y=df['number of floors'])
```

## Joinplot

```python
sns.jointplot(data = df,x = 'number of floors',y = 'Number of schools nearby')
```

##Line Plot

```python
plt.plot(df['Area of the basement'],df['Area of the house(excluding basement)'])
```

# **Multivariate Analysis**

##Pairplot

```python
X = df[['lot area','living area','number of floors','number of bedrooms','number of bathrooms','Price']]
sns.pairplot(X,dropna=True)
```

```python
df.drop(columns=['id','Date'],inplace=True)
sns.heatmap(df.corr(),annot=True)
```

# **Descriptive Statistics**

```python
a=df.groupby('number of floors')['Price'].median()
a.plot()
```
